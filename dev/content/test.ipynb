# Matplotlib Example

This notebook will demonstrate a basic example of using Matplotlib to create a simple line plot. We'll install the necessary packages, import them, and create a plot.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)                 # Compute the sine of each x point

# Create the plot
plt.figure()
plt.plot(x, y, label='Sine Wave')

# Add title and labels
plt.title('Basic Matplotlib Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Display the plot
plt.show()
```
